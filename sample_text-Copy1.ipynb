```python
import numpy as np
import pandas as pd
import chardet
```

```python
with open('spam.csv', 'rb') as file:
  encoding = chardet.detect(file.read())['encoding']
  df1 = pd.read_csv('spam.csv', encoding=encoding,usecols=[0, 1,])
  print(df1)
```

```python
df1
```

```python
df1['class'] = df1.v1.map({'ham':0, 'spam':1})
df1
```

```python
X= df1['v2'].to_numpy()
Y = df1['v1'].to_numpy()
y = Y.astype('int')
```

```python
print("X")
print(X)
print("y")
print(y)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
```

```python
vec.fit(X)
vec.vocabulary_
```

```python
vec = CountVectorizer(stop_words=['end','on','is'] )
vec.fit(X)
vec.vocabulary_
```

```python
print(vec.get_feature_names_out(input_features=None))
print(len(vec.get_feature_names_out(input_features=None)))
```

```python
X_transformed=vec.transform(X)
X_transformed
```

```python
X=X_transformed.toarray()
X
```

```python
pd.DataFrame(X, columns=vec.get_feature_names())
```

```python
df2 = pd.read_csv('test_text.csv') 
df2
```

```python
test_numpy_array = df2.to_numpy()
X_test = test_numpy_array[:,0]
y_test = test_numpy_array[:,1]
y_test = y_test.astype('int')
print("X_test")
print()
print("Y_test")
print(y_test)
```

```python
X_test_transformed=vec.transform(X_test)
X_test_transformed
```

```python
X_test=X_test_transformed.toarray()
X_test
```

```python
from sklearn.naive_bayes import MultinomialNB

mnb=MultinomialNB()

mnb.fit(X,y)

mnb.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score

y_prediction = mnb.predict(X_test)


acc = accuracy_score(y_test, y_prediction)


print("Accuracy:", acc)
```
